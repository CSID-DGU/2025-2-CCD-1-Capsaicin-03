```python
import whisper
import os
import csv

# PowerShell에서 ffmpeg 위치 확인 후 절대 경로 입력
os.environ["IMAGEIO_FFMPEG_EXE"] = r"C:\Users\betty\scoop\shims\ffmpeg.exe"

# 1️ Whisper 모델 불러오기 (CPU 사용)
model = whisper.load_model("medium")  # 필요하면 "small" 또는 "base"로 변경 가능

# 2️ 오디오 파일 폴더 경로 (Windows 절대 경로)
audio_folder = r"C:\Users\betty\cap_TTS\명령어 음성(소아, 유아)\Training\audio_sample"
csv_file = r"C:\Users\betty\cap_TTS\whisper_result.csv"

# 3️ 폴더 존재 여부 확인
if not os.path.exists(audio_folder):
    raise FileNotFoundError(f"오디오 폴더를 찾을 수 없습니다: {audio_folder}")

# 4️ 폴더 내 오디오 파일 목록 (확장자 필터, 소문자 통일)
audio_files = [f.strip() for f in os.listdir(audio_folder) if f.lower().endswith(('.mp3', '.wav', '.m4a'))]

if not audio_files:
    raise FileNotFoundError(f"폴더에 오디오 파일이 없습니다: {audio_folder}")

# 5️ CSV 파일 열기 및 헤더 작성
with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["VP", "Transcription"])

    # 6️ 각 오디오 파일 처리
    for filename in audio_files:
        audio_path = os.path.abspath(os.path.join(audio_folder, filename))
        print("Processing:", audio_path, "Exists:", os.path.exists(audio_path))

        # 7️ Whisper로 음성 -> 텍스트 변환
        result = model.transcribe(audio_path)

        # 8️ CSV에 기록
        writer.writerow(["VP", result["text"]])
        print(f"Processed {filename}")

print("모든 오디오 파일 처리 완료!")
```

```python
import os
import csv
import whisper
import subprocess

# -------------------------------
# 1️⃣ ffmpeg 절대경로 지정 (정식 설치)
# -------------------------------
ffmpeg_path = r"C:\Users\betty\scoop\shims\ffmpeg.exe"  # ffmpeg.exe 경로 확인
os.environ["IMAGEIO_FFMPEG_EXE"] = ffmpeg_path

# ffmpeg 실행 확인
subprocess.run([ffmpeg_path, "-version"])

# -------------------------------
# 2️⃣ Whisper 모델 불러오기
# -------------------------------
model = whisper.load_model("base")  # medium, small 등 변경 가능

# -------------------------------
# 3️⃣ 오디오 파일 폴더 (단순 경로)
# -------------------------------
audio_folder = r"C:\Users\betty\audio1.wav"  # 모든 오디오 파일 여기로 이동
csv_file = os.path.join(audio_folder, "whisper_result.csv")

# -------------------------------
# 4️⃣ 오디오 파일 목록
# -------------------------------
audio_files = [f for f in os.listdir(audio_folder) if f.lower().endswith(('.wav', '.mp3', '.m4a'))]

if not audio_files:
    raise FileNotFoundError("폴더에 오디오 파일이 없습니다!")

# -------------------------------
# 5️⃣ CSV 파일 작성
# -------------------------------
with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["VP", "Transcription"])

    for filename in audio_files:
        audio_path = os.path.abspath(os.path.join(audio_folder, filename))
        print("Processing:", audio_path, "Exists:", os.path.exists(audio_path))

        # -------------------------------
        # 6️⃣ Whisper로 음성 -> 텍스트 변환
        # -------------------------------
        try:
            result = model.transcribe(audio_path)
            writer.writerow(["VP", result["text"]])
            print(f"Processed {filename}")
        except Exception as e:
            print(f"Failed {filename}: {e}")
```
